```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
raw_data = pd.read_csv("C:/Users/LENOVO/Downloads/1.03.+Dummies.csv")
raw_data
```

```python
#dummy => imitating with numbers
data = raw_data.copy()
```

```python
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
```

## REGRESSION


```python
y = data['GPA']
x1=data[['SAT','Attendance']]
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

# regression for attending

```python
x2=data['SAT']
plt.scatter(x2,y)
yhat = 0.6439+0.2226 + 0.0014*x2
plt.plot(x2,yhat,lw=5,c='green',label='Regression Line')
plt.xlabel('SAT',fontsize=25)
plt.ylabel('GPA',fontsize=25)
plt.show()
```

# regression for not attending

```python
x2=data['SAT']
plt.scatter(x2,y)
y_not_hat = 0.6439 + 0.0014*x2
plt.plot(x2,y_not_hat,lw=5,c='orange',label='Regression Line')
plt.xlabel('SAT',fontsize=25)
plt.ylabel('GPA',fontsize=25)
plt.show()
```

# Now plotting both on same 

```python
x2=data['SAT']
plt.scatter(x2,y)
yhat = 0.6439+0.2226 + 0.0014*x2
y_org_hat = 0.0017*x2 + 0.2750
y_not_hat = 0.6439 + 0.0014*x2
plt.plot(x2,yhat,lw=5,c='green',label='Regression Line')
plt.plot(x2,y_org_hat,lw=5,c='red',label='Regression Line')
plt.plot(x2,y_not_hat,lw=5,c='orange',label='Regression Line')
plt.xlabel('SAT',fontsize=25)
plt.ylabel('GPA',fontsize=25)
plt.show()
```
