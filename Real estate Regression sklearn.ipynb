```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
sns.set()
```

```python
data = pd.read_csv("C:/Users/LENOVO/Downloads/real_estate_price_size (1).csv")
data.head()
```

# Regression

```python
y = data['price']
x = data['size']
x.shape
```

```python
y.shape
```

```python
x_matrix = x.values.reshape(-1,1)
reg = LinearRegression()
reg.fit(x_matrix,y)
```

```python
reg.intercept_
```

```python
x_matrix
```

```python
reg.coef_
```

```python
reg.score(x_matrix,y) # R squared value
```

## Predictions

```python
new_data = pd.DataFrame(data=[750],columns=['size'])
```

```python
new_data
```

```python
reg.predict(new_data)
```

```python
yhat = reg.intercept_ + reg.coef_*x_matrix
plt.scatter(x,y)
plt.plot(x,yhat,lw=5,c='black',label='LinearRegression')
plt.xlabel('SIZE')
plt.ylabel('PRICE')
plt.show()
```
