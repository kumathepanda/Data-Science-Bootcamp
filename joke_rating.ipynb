```python
data = raw_data.copy()
data = data.drop('id',axis=1)
```

```python
data
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate, train_test_split
from surprise.accuracy import rmse
# Define a Reader object with the appropriate rating scale
reader = Reader(rating_scale=(1, 5))

# Load the data into a Surprise dataset
surprise_data = Dataset.load_from_df(data[['user_id', 'joke_id', 'Rating']], reader)

# Split the data into training and test sets
trainset, testset = train_test_split(surprise_data, test_size=0.2)

# Use the SVD algorithm for matrix factorization
model = SVD()

# Train the model on the training set
model.fit(trainset)

# Predict ratings for the test set
predictions = model.test(testset)

# Evaluate the model using RMSE
print("RMSE: ", rmse(predictions))

# Example prediction for a specific user and joke
user_id = 'some_user_id'
joke_id = 'some_joke_id'
predicted_rating = model.predict(user_id, joke_id)
print(f'Predicted rating for user {user_id} and joke {joke_id}: {predicted_rating.est}')
```
