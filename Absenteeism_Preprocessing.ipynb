```python
import pandas as pd
import numpy as np
```

```python
raw_data=pd.read_csv('Absenteeism_data.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
data =  data.drop('ID',axis=1)
data
```

```python
data.columns
```

```python
data['Reason for Absence'].unique()
```

```python
# Create bins and labels
bins = [0, 14, 17, 21, 28]
labels = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']

# Cut the data into bins
data['Group'] = pd.cut(data['Reason for Absence'], bins=bins, labels= labels)

# Create dummy variables
data_dummies = pd.get_dummies(data['Group'])

# Concatenate the original dataframe with the dummies
data = pd.concat([data, data_dummies], axis=1)

# Drop the temporary 'Reason_Group' column
data.drop(columns=['Group'], inplace=True)

data
```

```python
data_with_dummies = data.drop('Reason for Absence',axis=1)
data_with_dummies
```

```python
data_with_dummies['Reason_1'] = data_with_dummies['Reason_1'].astype(int)
data_with_dummies['Reason_2'] = data_with_dummies['Reason_2'].astype(int)
data_with_dummies['Reason_3'] = data_with_dummies['Reason_3'].astype(int)
data_with_dummies['Reason_4'] = data_with_dummies['Reason_4'].astype(int)
```

```python
data_with_dummies
```

```python
data_with_dummies['Date'] = pd.to_datetime(data_with_dummies['Date'], format='%d/%m/%Y')
```

```python
data_with_dummies['Month'] = data_with_dummies['Date'].dt.month
data_with_dummies['DayOfWeek'] = data_with_dummies['Date'].dt.dayofweek
```

```python
data_with_dummies
```

```python
data_without_date = data_with_dummies.drop('Date',axis=1)
```

```python
data_without_date
```

```python
data_without_date['Education'] = data_without_date['Education'].apply(lambda x : 0 if x==1 else 1)
```

```python
data_without_date
```

```python
data_without_date.columns
```

```python
cols =  ['Reason_1', 'Reason_2','Reason_3', 'Reason_4', 'Month', 'DayOfWeek','Transportation Expense', 'Distance to Work', 'Age','Daily Work Load Average', 'Body Mass Index', 'Education', 'Children','Pets', 'Absenteeism Time in Hours']
data_without_date = data_without_date[cols]
```

```python
data_without_date.to_csv('Df_preprocessed.csv')
```
