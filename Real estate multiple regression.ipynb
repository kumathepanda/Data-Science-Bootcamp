```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/LENOVO/Downloads/real_estate_price_size_year (1).csv")
data.head()
```

```python
y = data['price']
x= data[['size','year']]
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
def adjusted_R_squared(x,y):
    R2 = reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    return (1-(1-R2)*(n-1)/(n-p-1))
```

```python
adjusted_R_squared(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

## Feature Selection


```python
from sklearn.feature_selection import f_regression
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values.round(3)
```
