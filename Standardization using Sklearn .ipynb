```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/LENOVO/Downloads/1.02.+Multiple+linear+regression (3).csv")
data
```

```python
y=data['GPA']
x = data[['SAT','Rand 1,2,3']]
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

```python
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame(data = ['Bias','SAT','Rand 1,2,3'],columns=['Features'])
reg_summary['weights']=reg.intercept_,reg.coef_[0],reg.coef_[1]
reg_summary
```

# Predicting


```python
new_data = pd.DataFrame(data = [[1700,1],[1800,2]],columns=['SAT','Rand 1,2,3'])
new_data
```

```python
reg.predict(new_data.values)
```

```python
new_data_scaled = scaler.transform(new_data)
```

```python
reg.predict(new_data_scaled)
```
