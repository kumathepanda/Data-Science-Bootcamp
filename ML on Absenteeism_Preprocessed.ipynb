# Create Logistic regression for predicting absenteeism

## importing libraries


```python
import numpy as np
import pandas as pd
```

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
data_preprocessed.head()
```

## create the targets

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
# 3 hours is our cutoff line above it is absent and less than that is not 
targets= np.where(data_preprocessed['Absenteeism Time in Hours']>data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
```

```python
targets
```

```python
targets.sum()/targets.shape[0] #roughly balanced
```

```python
data_preprocessed['Excessive Absenteesism'] = data_preprocessed['Absenteeism Time in Hours'].apply(lambda x : 1 if x>3.0 else 0)
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours','Day of Week','Daily Work Load Average','Distance to Work'],axis=1)
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

```python
from sklearn.preprocessing import StandardScaler

class CustomScaler:
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.columns = columns
        self.mean_ = None

    def fit(self, X, y=None):
        self.mean_ = X[self.columns].mean()
        self.scaler.fit(X[self.columns], y)
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]

    def fit_transform(self, X, y=None, copy=None):
        return self.fit(X, y).transform(X, y, copy)
```

```python
unscaled_inputs.columns.values
```

```python
columns_to_omit = ['Reason_group1', 'Reason_group2', 'Reason_group3', 'Reason_group4','Education']
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit]
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale)
```

```python
scaled_inputs = absenteeism_scaler.fit_transform(unscaled_inputs)
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

## finding the intercept and coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['feature name'],data=feature_name)
summary_table['coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

## Interpreting the coefficients

```python
# larger the coefficient more the weight (more important feature)
summary_table['odds_ratio'] = np.exp(summary_table.coefficient)
```

```python
summary_table.sort_values('odds_ratio',ascending=False)
```

## Testing the model
