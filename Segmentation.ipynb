```python
import pandas as pd
import numpy as np
import seaborn as sns 
sns.set()
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("C:/Users/LENOVO/Downloads/3.12.+Example.csv")
data
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x =data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['clusters_pred'] = kmeans.fit_predict(x)
plt.scatter(data['Satisfaction'],data['Loyalty'],c=clusters['clusters_pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('number of clusters')
plt.ylabel('wcss')
```

```python
kmeans_new = KMeans(4)
kmeans_new.fit(x_scaled)
clusters_new = x.copy()
clusters_new['pred'] = kmeans_new.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(data['Satisfaction'],data['Loyalty'],c=clusters_new['pred'],cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
